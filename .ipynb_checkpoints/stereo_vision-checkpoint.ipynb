```python
import numpy as np
import scipy as sp
import imageio as im
import matplotlib.pyplot as plt
```

```python
def img_hConv(img1,img2):
    temp = np.zeros(img1.shape[1]*2)
    for i in np.arange(1,img1.shape[1]*2):
        if (i <=img1.shape[1]):
            temp[i] = sum(sum(img1[:,0:i]*img2[:,0:i]))
        else:
            temp[i] = sum(sum(img1[:,i-img1.shape[1]:img1.shape[1]]*img2[:,i-img1.shape[1]:img1.shape[1]]))
    return temp
```

```python
gray  = lambda img: 0.21*img[:,:,0]+0.72*img[:,:,0]+0.07*img[:,:,0]
get_nbhd = lambda x,y,nbhd,img: img[x-nbhd:x+nbhd,y-nbhd:y+nbhd]
img_right = gray(im.imread("right.png"))
img_left = gray(im.imread("left.png"))
d_k = 5; #Camera Distance
```

```python
plt.subplot(1,2,1)
plt.imshow(img_left,cmap="gray")
plt.subplot(1,2,2)
plt.imshow(img_right,cmap="gray")
```
